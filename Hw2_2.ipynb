10)Створіть нову таблицю python_data, в якій будуть тільки рядки зі спеціалістами, які вказали мову програмування Python.

```python
df_1 = pd.read_csv("2017_jun_final.csv")
python_data = df_1[df_1['Язык.программирования'] == 'Python']
python_data
```

12)Використовуючи метод groupby, виконайте групування за стовпчиком "Посада".

```python
grouped_data = python_data.groupby('Должность')
```

13)Створіть новий DataFrame, де для згрупованих даних за стовпчиком "Посада", виконайте агрегацію даних за допомогою методу agg і знайдіть мінімальне та максимальне значення у стовпчику "Зарплата.в.місяць".

```python
salaries = grouped_data['Зарплата.в.месяц'].agg(['min', 'max'])
salaries
```

14)Створіть функцію fill_avg_salary, яка повертатиме середнє значення заробітної плати на місяць. Використовуйте її для методу apply та створіть новий стовпчик "avg".

```python
def fill_avg_salary(row):
    return row.mean()

salaries['avg'] = salaries.apply(fill_avg_salary, axis=1)
```

15)Створіть описову статистику за допомогою методу describe для нового стовпчика.

```python
salaries["avg"].describe()
```

16)Збережіть отриману таблицю в CSV файл.

```python
salaries.to_csv('salaries.csv')
```

**Візуалізація:**

```python
import matplotlib.pyplot as plt
```

```python
data = df[df['Язык.программирования'] == 'Python']
job_data = data['Должность'].value_counts()
plt.title('Розподіл програмістів за посадами')
plt.pie(job_data, labels=job_data.index, autopct="%.2f%%")

plt.show()
```

```python
import seaborn as sns

data = df[df['Язык.программирования'] == 'Python']
sns.scatterplot(data.sort_values(by=['Зарплата.в.месяц'], ascending=True), x='Должность', y='Зарплата.в.месяц', hue='Язык.программирования')
plt.title('Рівень зарплати для Пітоністів в розрізі посади')
plt.xticks(rotation=90)
plt.legend(bbox_to_anchor=(1.05, 0.5), loc='center left')
plt.show()
```

```python
study_data = df.groupby('Образование')
study_data = study_data.agg({"Зарплата.в.месяц": 'mean'})
sns.barplot(data=study_data, y='Зарплата.в.месяц', x=study_data.index)
plt.title('Середня зарплата програмістів, залежно від освіти')
plt.xticks(rotation=90)
plt.show()
```
