```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from datetime import datetime
```

```python
xf = pd.read_csv("Balance model data extended event.csv", delimiter=";")
xf.head()
```

```python
yf = pd.read_csv("Y Balances.csv", delimiter=";")
yf.head()
```

```python
x_train = np.array(xf)
y_train = np.array(yf)
print(f"x_train = {x_train}")
print(f"y_train = {y_train}")
```

```python
a = []
with open('Balance model data extended event.csv', 'r') as f:
  b = f.readlines()
  a.append(b[0].split(';'))
  for i in b[1:]:
    parts = i.split(';')
    a.append([datetime.strptime(parts[0],'%m/%d/%Y').strftime('%j'), parts[1].replace(' ', ''), *parts[2:]])
with open('Balance model data float.csv', 'w') as f:
  for i in a:
    f.write(';'.join(i))
```

```python
b = []
with open('Y Balances.csv', 'r') as f:
  c = f.readlines()
  b.append(c[0].split(';'))
  for i in c[1:]:
    parts = i.split(';')
    b.append([datetime.strptime(parts[0],'%m/%d/%Y').strftime('%j'), parts[1].replace(' ', '')])
with open('Y Balances data float.csv', 'w') as f:
  for i in b:
    f.write(';'.join(i))
```

```python
x1f = pd.read_csv("Balance model data float.csv", delimiter=";")
y1f = pd.read_csv("Y Balances data float.csv", delimiter=";")
x1f.head()
y1f.head()
```

```python
x1_train = np.array(x1f)
y1_train = np.array(y1f)
print(f"x1_train = {x1_train}")
print(f"y1_train = {y1_train}")
```

```python
linear_model = LinearRegression()
linear_model.fit(x1_train, y1_train)
```

```python
b = linear_model.intercept_
w = linear_model.coef_
print(f"w = {w:}, b = {b:0.2f}")
```

```python
print(f"Prediction on training set:\n {linear_model.predict(x1_train)[:6]}" )
print(f"prediction using w,b:\n {(x1_train @ w + b)}")
print(f"Target values \n {y1_train[:6]}")

x = np.array([213, 363000000,750, 4, 1, 9]).reshape(-1,6)
x_predict = linear_model.predict(x)[0]
print(f" prediction = ${x_predict:0.2f}")
```
