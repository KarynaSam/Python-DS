```python
import numpy as np
import pandas as pd
```

Прочитайте дані за допомогою методу read_html з таблиці "Коефіцієнт народжуваності в регіонах України (1950—2019)"

```python
data_url = "https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%81%D0%B5%D0%BB%D0%B5%D0%BD%D0%BD%D1%8F_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D0%B8"
match_text = "Коефіцієнт народжуваності в регіонах України"
data = pd.read_html(data_url, match=match_text)
df = data[0]
display(df)
```

3)Замініть у таблиці значення "—" на значення NaN

```python
df.replace('—', np.nan, inplace=True)
display(df)
```

4) Порахуйте, яка частка пропусків міститься в кожній колонці (використовуйте методи isnull та sum)

6)Замініть типи нечислових колонок на числові. Підказка - це колонки, де знаходився символ "—"

```python
df['2014'] = df['2014'].fillna(0).astype('int64')
df['2019'] = df['2019'].fillna(0).astype('int64')
print(df.dtypes)
df
```

7)Порахуйте, яка частка пропусків міститься в кожній колонці (використовуйте методи isnull та sum)

```python
df.isnull().sum()
```

8)Видаліть з таблиці дані по всій країні, останній рядок таблиці

```python
df = df.drop(len(df)-1)
display(df)
```

9)Замініть відсутні дані в стовпцях середніми значеннями цих стовпців (метод fillna)

```python
df[df['2014']>df['2014'].mean()]
```

10)Отримайте список регіонів, де рівень народжуваності у 2019 році був вищим за середній по Україні

```python
mean_2019 = df["2019"].mean()

df[df["2019"]> mean_2019]["Регіон"].tolist()
```

11)У якому регіоні була найвища народжуваність у 2014 році?

```python
df.loc[df['2014'] == df['2014'].max()].iloc[:, 0]
```

12)Побудуйте стовпчикову діаграму народжуваності по регіонах у 2019 році

```python
import matplotlib.pyplot as plt
```

```python
plt.bar(df['Регіон'], df['2019'], color = "indigo")
plt.xlabel("Регіон", fontsize=10, color="indigo")
plt.ylabel("Рівень народжуваності", fontsize=10, color="indigo")
plt.title("Діаграма народжуваності по регіонах у 2014 році", fontsize=15, color="indigo")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
